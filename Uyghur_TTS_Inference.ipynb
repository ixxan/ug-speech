# Facebook MMS TTS

```python
import torch
if torch.cuda.is_available():
  device = "cuda"
else:
  device = 'cpu'
device
```

```python
# !pip install --upgrade transformers accelerate
```

```python
from transformers import VitsModel, AutoTokenizer
import torch

model_mms = VitsModel.from_pretrained("facebook/mms-tts-uig-script_arabic")
tokenizer_mms = AutoTokenizer.from_pretrained("facebook/mms-tts-uig-script_arabic")
```

```python
from IPython.display import Audio
text = "مەن ئۇيغۇرچە سۆزلىيەلەيمەن."
inputs = tokenizer_mms(text, return_tensors="pt").to(device)
model_mms.to(device)

with torch.no_grad():
    output = model_mms(**inputs).waveform.cpu()

Audio(output, rate=model_mms.config.sampling_rate)
```

# MMS TTS Fine tuned

```python
from transformers import AutoTokenizer, VitsModel
import torch

# Load the tokenizer and model
model_name = "ixxan/mms-tts-uig-script_arabic-UQSpeech"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model = VitsModel.from_pretrained(model_name)
```

```python
from IPython.display import Audio
text = "بۇ پۇتبول مۇسابىقىسىنىڭ ئاخىرلىشىشى"
inputs = tokenizer(text, return_tensors="pt").to(device)
model.to(device)

with torch.no_grad():
    output = model(**inputs).waveform.cpu()

Audio(output, rate=model.config.sampling_rate)
```

```python
from IPython.display import Audio
text = "بۇ پۇتبول مۇسابىقىسىنىڭ ئاخىرلىشىشى"
inputs = tokenizer(text, return_tensors="pt").to(device)
model.speaking_rate = 1.2
model.to(device)

with torch.no_grad():
    output = model(**inputs).waveform.cpu()

Audio(output, rate=model.config.sampling_rate)
```

```python
# Import the necessary library and load the model
from parallel_wavegan.utils import load_model
from espnet2.bin.tts_inference import Text2Speech

vocoder_checkpoint="parallelwavegan_male2_checkpoint/checkpoint-400000steps.pkl" ### specify vocoder path
vocoder = load_model(vocoder_checkpoint).to(device).eval()
vocoder.remove_weight_norm()

### specify path to the main model(transformer/tacotron2/fastspeech) and its config file
config_file = "exp/tts_train_raw_char/config.yaml"
model_path = "exp/tts_train_raw_char/train.loss.ave_5best.pth"

text2speech = Text2Speech(
    config_file,
    model_path,
    device=device, ## if cuda not available use cpu
    ### only for Tacotron 2
    threshold=0.5,
    minlenratio=0.0,
    maxlenratio=10.0,
    use_att_constraint=True,
    backward_window=1,
    forward_window=3,
    ### only for FastSpeech & FastSpeech2
    speed_control_alpha=1.0,
)
text2speech.spc2wav = None  ### disable griffin-lim
```

```python
from turkicTTS_utils import normalization
from IPython.display import Audio
fs = 23000
```

```python
text = "rehmet sizge."
### available options are azerbaijani, bashkir, kazakh, kyrgyz, sakha, tatar, turkish, turkmen, uyghur, uzbek
lang = "uyghur"

text = normalization(text, lang)
with torch.no_grad():
    c_mel = text2speech(text)['feat_gen']
    wav = vocoder.inference(c_mel)
output = wav.view(-1).cpu().numpy()


Audio(output, rate=22050)
```
